```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import datetime
from datetime import date, timedelta
```

```python
today = date.today()
d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=730)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
```

```python
data = yf.download('BTC-USD', start=start_date, end=end_date, progress=False)
```

```python
from autots import AutoTS
```

```python
train_data1 = data.iloc[:-10] 
test_data1 = data.iloc[-10:]
```

```python
train_data2 = data.iloc[:-20] 
test_data2 = data.iloc[-20:]
```

```python
train_data3 = data.iloc[:-30] 
test_data3 = data.iloc[-30:]
```

```python
model1 = AutoTS(forecast_length=10, frequency='infer', ensemble='simple')
model1 = model1.fit(train_data1, date_col='Date', value_col='Close', id_col=None)
```

```python
model2 = AutoTS(forecast_length=20, frequency='infer', ensemble='simple')
model2 = model2.fit(train_data2, date_col='Date', value_col='Close', id_col=None)
```

```python
model3 = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model3 = model3.fit(train_data3, date_col='Date', value_col='Close', id_col=None) 
```

```python
prediction1 = model1.predict()
forecast1 = prediction1.forecast
```

```python
prediction2 = model2.predict()
forecast2 = prediction2.forecast
```

```python
prediction3 = model3.predict()
forecast3 = prediction3.forecast
```
